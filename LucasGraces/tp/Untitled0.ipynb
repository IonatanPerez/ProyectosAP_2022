```python
y = data['Precio']
```

```python
y.head(5)
```

```python
x = data.drop(columns = ['Precio']) 
```

```python
data.head(7)
```

```python
X = data   #variables predictora
# y variable a predecir
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=138)  #por default 25% de test
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
arbol = DecisionTreeRegressor(max_depth=10, )
```

```python
arbol.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error

y_pred_train = arbol.predict(X_train)
mse_train_arbol = mean_squared_error(y_train, y_pred_train)
mse_train_arbol
```

```python
y_pred_test = arbol.predict(X_test)
mse_train_arbol = mean_squared_error(y_test, y_pred_test)
mse_train_arbol
```

```python
len(y_pred_arbol)
```

```python
from sklearn.metrics import confusion_matrix

matriz_arbol = confusion_matrix(y_test, y_pred_arbol)
sns.heatmap(matriz_arbol, annot=True)
plt.xlabel("Etiquetas predichas")
plt.ylabel("Etiquetas reales")
```
