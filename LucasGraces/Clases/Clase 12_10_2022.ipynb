```python
import geopandas as gpd
```

```python
barrios = gpd.read_file("barrios.geojson")
puntosverdes = gpd.read_file("puntos-verdes.geojson")
```

```python
puntosverdes.plot()
```

```python
barrios.plot()
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1,figsize=(10,10))
barrios.plot(ax=ax, edgecolor='black', facecolor='none')
puntosverdes.plot(ax=ax,color='green')
#barrios.plot(ax=ax, edgcolor = 'k', facecolor='none')
puntosverdes.apply(lambda x: ax.annotate(s=x['id'], xy=x['geometry'].coords[0]),fontsize=20, axis=1)
```
