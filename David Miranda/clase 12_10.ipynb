```python
import geopandas as gpd
```

```python
ptos_verdes = gpd.read_file('puntos-verdes.geojson')
barrios = gpd.read_file('barrios.geojson')
```

```python
ptos_verdes.head()
```

```python
ptos_verdes.plot()
```

```python
barrios.plot()
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1, figsize = (20,20))
barrios.plot(ax = ax, edgecolor = 'black', facecolor = 'none')
ptos_verdes.plot(ax = ax, color = 'green')
ptos_verdes.apply(lambda x: ax.annotate(s = x['id'], xy = x['geometry'].coords[0]), axis = 1)
```
