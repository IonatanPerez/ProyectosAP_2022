```python
import pandas as pd
```

```python
data = pd.read_csv("datasets_modif.csv")
data
```

```python
drop_tablas = ["title", "description", "price_period", "Unnamed: 0", "start_date", "end_date", "created_on"]
data.drop(columns = drop_tablas, inplace = True)
```

```python
dict_renombrar= {"l3":"Barrio", "l4":"Zona", "rooms":"Habitaciones", "price":"Precio", "property_type":"Tipo_propiedad"}
data.rename(columns=dict_renombrar, inplace =True)
```

```python
data
```

```python
data = pd.get_dummies(data,  drop_first=True)
data
```

```python
data["Habitaciones"].loc[data["Habitaciones"].isnull()] = "Sin datos"
```

```python
#data = pd.get_dummies(data,  drop_first=True)
data
```

```python
import numpy as np
dict_amb = {'Sin datos':'Sin datos',1:'monoambiente',2:'dos amb',3:'3 amb'}
```

```python
data['Habitaciones'] = data['Habitaciones'].map(dict_amb).fillna('4+ amb')
```

```python
data
```

```python
data = pd.get_dummies(data,  drop_first=True)
data
```

```python
data_limpia = data.dropna()
data_limpia
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
```

```python
X = data_limpia.drop(columns = 'Precio')   #variables predictora
y = data_limpia["Precio"]   #variable a predecir
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=138)
```

```python
arbol = DecisionTreeRegressor()
```

```python
arbol.fit(X_train, y_train)
```

```python
y_pred_train = arbol.predict(X_train)
mse_train_arbol = mean_squared_error(y_train, y_pred_train, squared = False)
mse_train_arbol
```

```python
y_pred_test = arbol.predict(X_test)
mse_test_arbol = mean_squared_error(y_test, y_pred_test, squared = False)
mse_test_arbol
```

```python
arbol = DecisionTreeRegressor(max_depth = 10)
arbol.fit(X_train, y_train)
y_pred_train = arbol.predict(X_train)
mse_train_arbol = mean_squared_error(y_train, y_pred_train, squared = False)
print (mse_train_arbol)
y_pred_test = arbol.predict(X_test)
mse_test_arbol = mean_squared_error(y_test, y_pred_test, squared = False)
print (mse_test_arbol)
```
